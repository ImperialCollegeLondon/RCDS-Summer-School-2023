```python
import numpy as np

data_points_time = np.arange(0, 21)
data_points_temperature = np.array([0, 1, 2, 5, 10, 20, 18, 14, 12, 11, 10, 10, 11, 13, 15, 17, 19, 20.5, 19, 17, 14])

actual_time = np.arange(3, 7.01, 0.1)
actual_temperature = np.array([5, 
5.4, 
5.8, 
6.2, 
6.8, 
7.4, 
7.9, 
8.4, 
8.9, 
9.4, 
10, 
10.7, 
11.5, 
12.4, 
13.4, 
14.5, 
15.7, 
17, 
18.4, 
19.2, 
20, 
20.5, 
20.8, 
21, 
21.1, 
21, 
20.8, 
20.2,
19.6,  
18.8, 
18, 
17.4, 
16.8, 
16.4, 
16, 
15.6, 
15.2, 
14.9, 
14.6, 
14.3,
14
])

def quadratic_function(x):
    return -6 * x ** 2 + 64 * x -150

fit_time = np.arange(4, 6.01, 0.1)
fit_temperature = quadratic_function(fit_time)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(data_points_time, data_points_temperature)

ax.set_xlabel('Time (hours)')
ax.set_ylabel('Temperature (Celsius)')

fig.savefig('peak_full.png')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(data_points_time, data_points_temperature, label='Data Points')
ax.plot(actual_time, actual_temperature, label='Actual Temperature', color='red')

ax.set_xlabel('Time (hours)')
ax.set_ylabel('Temperature (Celsius)')
ax.set_xbound(3,7)

ax.legend()

fig.savefig('peak_closeup.png')
```

```python
import matplotlib.pyplot as plt



fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(data_points_time, data_points_temperature, label='Data Points')
ax.plot(actual_time, actual_temperature, label='Actual Temperature', color='red')
ax.plot(fit_time, fit_temperature, label='Quadratic Fit', color='green')

ax.set_xlabel('Time (hours)')
ax.set_ylabel('Temperature (Celsius)')
ax.set_xbound(3,7)

ax.legend()

fig.savefig('peak_fit.png')
```
